VARIABLE : 
A variable is just a label you stick on a value so you can use it later. Like writing a
name on a sticky note and slapping it on a box.

--- 
VARIABLE ASSIGNMENT :
'=' means "store this value with this label". It's not math equality — it's just saving a value.
e.g. 
* name = "Tariq"
* age = 25
* is_admin = True

```python
name = "Appu"
age = 23
present = False 

print(name , age , present )
```

--- 
NAMING RULES :
* variables and function - snake_case e.g. user_name , get_data
* Constants - UPPER_CASE  e.g. MAX_RETRIES = 3
* Classes - PascalCase e.g. class AzureClient

A few hard rules Python enforces:
* Can't start with a number → 2name is invalid
* No spaces → use _ instead
* Never use built-in names like list, str, id as variable names — you'll break Python's own tools

--- 
CHECK WHAT TYPE A VARIABLE IS : type() function - useful when debugging 


```python
name = "Appu"
age = 23
marks = 75.4
present = False 
backlogs = None 

print(type(name))
print(type(age))
print(type(marks))
print(type(present))
print(type(backlogs))
```

--- 
BUIL-IN DATATYPES :
* int (integer)- whole numbers (25 , -6 , 0 )
* float (float)- decimal numbers (2.45 , -8.9 , 0.54)
* str (string)- text ("hello" , 'hi')
* bool (boolean)- true or false (True , False)
* NoneType - no value at all (None)

```python
age = 25          # int
price = 9.99      # float
name = "Appu"    # str
active = True     # bool
result = None     # NoneType
```

---
CONVERTING BETWEEN TYPES (TYPE CASTING) : 
Sometimes we get a number as text and need to do math on it — that's where conversion comes in.

```python
# str to int
age = int("25")        # "25" -> 25

# str to float
price = float("9.99")  # "9.99" -> 9.99

# int to str
label = str(100)       # 100 -> "100"

# anything to bool
bool(1)    # True
bool(0)    # False
bool("")   # False  — empty string is False
bool("hi") # True   — anything with content is True
```

--- 
FALSY VALUES :

These all count as False when used in conditions

Everything else is True. We'll use this constantly in if statements.

```python
bool(0)     # False
bool(0.0)   # False
bool("")    # False
bool(None)  # False
bool([])    # False — empty list
```

* None is not zero and not False

None means "there is no value here at all" — it's not empty, it's absent.

```python
x = None

x == False   # False — they are not the same
x == 0       # False — not the same

# correct way to check for None
x is None    # True
```
