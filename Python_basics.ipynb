WHAT IS A PROGRAMMING LANGUAGE ? :
* A programming language is a formal way to give instructions to a computer.
* It helps humans communicate with machines using readable commands.
* Used to create software, apps, websites, AI models, games, and more.
* Examples: Python, Java, C, C++, JavaScript.
    
In simple words: “A programming language tells the computer what to do, step by step.”

COMPILER VS INTERPRETER :

| Compiler                       | Interpreter                      |
| ------------------------------ | -------------------------------- |
| Translates the whole program   | Translates line by line          |
| Faster execution               | Slower execution                 |
| Generates executable file      | No executable generated          |
| Errors shown after compilation | Errors shown immediately         |
| Better performance             | Better for testing and debugging |

WHAT IS PYTHON AND WHY WE NEED IT ?:
* Python is a high-level, interpreted, General Purpose programming language used to build software, websites, games, apps, and more.
* It is known for its simple syntax, which looks like everyday English, making it easy to learn for beginners.
* Developed by Guido van Rossum and released in 1991.

---
KEY FEATURES OF PYTHON :
* Easy to learn and use - similar to english 
* Interpreted language - runs line by line , the error doesnt show up unless that particular line is run 
* High level Language - humans understandable 
* Portable - can work on different platforms 
* Free and Open-Source - anyone can use it for free , contribute to it and redistribute it 
* Extensive Libraries - a large suite of libraries for different things 
* Object Oriented and Functional - suppourts object oriented programming and functions 
* Dynamic Typing(Implicit programming) - no need to declare the variables and their types before using , we could declare it when we use it and python automatically sets the type 
* Large community support - large number of actively contributing developers and also proffessionals who suppourt debugging

PYTHON SYNTAX:
* Clean and readable like English
* No need for semicolons ; or curly brackets { }
* Indentation defines code blocks (very important!)
