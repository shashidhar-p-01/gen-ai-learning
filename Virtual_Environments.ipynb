VIRTUAL ENVIRONMENTS :
Think of venv like a separate room for each project. Every project gets its own Python packages so they don't mix with each other.

---
TO CREATE A VIRTUAL ENVIRONMENT :
* $ python3 -m venv myvenv (for linux)
  
* $ python -m venv myvenv (for windows)

---
TO ACTIVATE VIRTUAL ENVIRONMENT :
* $ source ./myvenv/bin/activate (for linux )

* $ myvenv/Scripts/activate (for windows)
  
When it's active you'll see (venv) at the start of your terminal line. That means you're inside it.

---
TO DEACTIVATE VIRTUAL ENVIRONMENT :
* $ deactivate
